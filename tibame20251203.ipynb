````python
# https://sites.google.com/dcjh.tn.edu.tw/python/l06%E9%BE%9C%E5%9C%96%E5%89%B5%E4%BD%9C
import ColabTurtle.Turtle as turtle
import time

turtle.initializeTurtle()


turtle.speed(8)
init_width = 40
shape_count = 5
shape_angle = 360 / shape_count

prev_angle = 0

points_1 = [0] * shape_count
points_2 = [0] * shape_count
i = 0
while i < shape_count:
    turtle.forward(init_width)
    turtle.right(shape_angle)
    turtle.forward(init_width)
    # 儲存頂點
    points_1[i] = turtle.pos()
    turtle.right(180-shape_angle)
    turtle.forward(init_width)
    # 儲存下去的點
    points_2[i] = turtle.pos()
    turtle.right(shape_angle)
    turtle.forward(init_width)
    turtle.left(180)
    # 計算下一次的菱形角度
    # 2 * 原本角度 - 上一次角度
    i = i + 1

for i in range(shape_count):
    time.sleep(2)
    new_points_1 = [0] * shape_count
    new_points_2 = [0] * shape_count
    i = 0
    while i < shape_count:
        p1_x, p1_y = points_1[i]
        turtle.penup()
        turtle.goto(p1_x, p1_y)
        turtle.pendown()
        nex_p1_x, nex_p1_y = points_1[(i+1)%len(points_1)]
        p1_nex_center_x, p1_nex_center_y = (p1_x + nex_p1_x) / 2.0, (p1_y + nex_p1_y) / 2.0
        p2_x, p2_y = points_2[i]
        delta_x, delta_y = p1_nex_center_x - p2_x, p1_nex_center_y - p2_y
        # !
        nex_goto_x, nex_goto_y = p2_x + 2 * delta_x, p2_y + 2 * delta_y
        turtle.goto(nex_goto_x, nex_goto_y)
        new_points_1[i] = turtle.pos()
        turtle.goto(nex_p1_x, nex_p1_y)
        new_points_2[i] = turtle.pos()
        i = i + 1
    points_1 = new_points_1
    points_2 = new_points_2
````
