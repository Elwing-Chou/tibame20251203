````python
def func(s):
    s_spl = s.split(" / ")
    if len(s_spl) == 1:
        area = "-"
        genre = s_spl[0]
    else:
        area = s_spl[0]
        genre = s_spl[1]
    ans = {"area":area, "genre":genre}
    return pd.Series(ans)

df[["area", "genre"]] = df["area_genre"].apply(func)
df
````

````python
def func(s):
    ans = {
        "dinner(low)":float("inf"),
        "dinner(high)":float("inf")
    }
    s = s.replace("JPY ", "")
    s = s.replace(",", "")
    if s == "-":
        pass
    elif s.startswith("-"):
        ans["dinner(high)"] = s.replace("- ", "")
    # guess
    elif s.endswith("-"):
        ans["dinner(low)"] = s.replace(" -", "")
    else:
        low, high = s.split(" - ")
        ans["dinner(high)"] = high
        ans["dinner(low)"] = low
    return pd.Series(ans)

df[["dinner(low)", "dinner(high)"]] = df["dinner_price"].apply(func).astype("float")
df
````

````python
fil1 = df["dinner_price"] != "-"
fil2 = df["dinner(high)"] <= 10000
fil = fil1 & fil2
df[fil]
````

````python
# 南港(06:15)/台北(06:26)
import json
def func(s):
    # print(s)
    s = s.replace("'", '"')
    s = s.replace("True", "true")
    s = s.replace("False", "false")

    ans = ""
    stations = json.loads(s)
    for stat in stations:
        if stat["Show"] == True:
            ans = ans + stat["StationName"] + "(" + stat["DepartureTime"] + ")/"
    return ans
df["Stop_Station"] = df["StationInfo"].apply(func)
# df = df.drop("StationInfo", axis=1)
df
````
