````python
def func(s):
    s_spl = s.split(" / ")
    if len(s_spl) == 1:
        area = "-"
        genre = s_spl[0]
    else:
        area = s_spl[0]
        genre = s_spl[1]
    ans = {"area":area, "genre":genre}
    return pd.Series(ans)

df[["area", "genre"]] = df["area_genre"].apply(func)
df
````

````python
def func(s):
    ans = {
        "dinner(low)":float("inf"),
        "dinner(high)":float("inf")
    }
    s = s.replace("JPY ", "")
    s = s.replace(",", "")
    if s == "-":
        pass
    elif s.startswith("-"):
        ans["dinner(high)"] = s.replace("- ", "")
    # guess
    elif s.endswith("-"):
        ans["dinner(low)"] = s.replace(" -", "")
    else:
        low, high = s.split(" - ")
        ans["dinner(high)"] = high
        ans["dinner(low)"] = low
    return pd.Series(ans)

df[["dinner(low)", "dinner(high)"]] = df["dinner_price"].apply(func).astype("float")
df
````

````python
fil1 = df["dinner_price"] != "-"
fil2 = df["dinner(high)"] <= 10000
fil = fil1 & fil2
df[fil]
````
