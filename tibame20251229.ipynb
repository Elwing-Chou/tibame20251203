````python
def func(n):
    return round(n * 2)

df["rating"].apply(func)
````

````python
def func(s):
    s_spl = s.split(" / ")
    if len(s_spl) == 1:
        area = "-"
        genre = s_spl[0]
    else:
        area = s_spl[0]
        genre = s_spl[1]
    ans = {"area":area, "genre":genre}
    return pd.Series(ans)

df[["area", "genre"]] = df["area_genre"].apply(func)
df
````
