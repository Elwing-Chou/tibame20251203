````python

# 五子棋
# 1. 先把棋盤畫出來
# 2. 落子(落到最相鄰的位置)
# 3. 判斷落子是否五子
import pygame as pg

PLAYER_UNSET = 0
PLAYER_1 = 1
PLAYER_2 = 2

def check_win(board, pos_i, pos_j):
    player = board[pos_i][pos_j]
    count = 0
    # maxstep = min(4, board_line_count-pos_j-1)
    # for j_offset in range(maxstep):
    #     nex = board[pos_i][pos_j+j_offset]
    #     if nex == player:
    #         count = count + 1
    #     else:
    #         break
    offset = 1
    while True:
        new_i, new_j = pos_i, pos_j + offset
        if 0 <= new_i < len(board) and 0 <= new_j < len(board[0]):
            nex = board[new_i][new_j]
            if nex == player:
                count = count + 1
            else:
                break
        else:
            break
        offset = offset + 1
    offset = 1
    while True:
        new_i, new_j = pos_i, pos_j - offset
        if 0 <= new_i < len(board) and 0 <= new_j < len(board[0]):
            nex = board[new_i][new_j]
            if nex == player:
                count = count + 1
            else:
                break
        else:
            break
        offset = offset + 1
    if count >= 4:
        return True
    return False

# pygame初始化
pg.init()
# 設定視窗
width, height = 640, 640
board_line_count = 15
board = [[PLAYER_UNSET] * 15 for i in range(15)]
width_inter = width / (board_line_count + 1)
height_inter = height / (board_line_count + 1)

# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("五子棋")
# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])
# 開始畫棋盤
i = 0
while i < board_line_count:
    pg.draw.line(bg,
                 [50, 50, 50],
                 [width_inter, height_inter*i+height_inter],
                 [width-width_inter, height_inter*i+height_inter],
                 2)
    i = i + 1

i = 0
while i < board_line_count:
    pg.draw.line(bg,
                 [50, 50, 50],
                 [width_inter*i+width_inter, height_inter],
                 [width_inter*i+width_inter, height-height_inter],
                 2)
    i = i + 1

# 把某個圖層貼到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

game_round = 0
running = True
while running:
    for event in pg.event.get():
        if event.type == pg.MOUSEBUTTONUP:
            x, y = pg.mouse.get_pos()
            print(x, y)
            # x:0-640 x_scale:0-16(float)
            x_scale = x / width_inter
            y_scale = y / height_inter
            x_scale_round = round(x_scale)
            y_scale_round = round(y_scale)
            y2i, x2j = y_scale_round - 1, x_scale_round - 1

            if (1 <= x_scale_round <= board_line_count and
                1 <= y_scale_round <= board_line_count and
                board[y2i][x2j] == PLAYER_UNSET):
                x_pos = x_scale_round * width_inter
                y_pos = y_scale_round * height_inter
                if game_round % 2 == 0:
                    player_color = [0, 0, 0]
                    player = PLAYER_1
                else:
                    player_color = [255, 255, 255]
                    player = PLAYER_2

                pg.draw.circle(bg,
                               player_color,
                               [x_pos, y_pos],
                               min(width_inter/2, height/2),
                               0)
                board[y2i][x2j] = player
                game_round = game_round + 1
                # 把某個圖層貼到上一層
                screen.blit(bg, [0, 0])
                # 對畫面進行更新(才會真的秀出來)
                pg.display.update()
                if check_win(board, y2i, x2j):
                    running = False
        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False
pg.quit()

````

````python
import random

def generate_ticket(lo=1, hi=49, count=6):
    ticket = set()
    while True:
        n = random.randint(lo, hi)
        ticket.add(n)
        if len(ticket) == count:
            break
    return ticket

prize = generate_ticket()
print(prize)

special = None
while True:
    special = random.randint(1, 49)
    if not special in prize:
        break
print(special)

count = 0
while True:
    my = generate_ticket()
    count = count + 1
    prize_numbers = my & prize
    score = 2 * len(prize_numbers)
    if special in my:
        score = score + 1
    if score >= 11:
        break

print(count, count * 50)
````
