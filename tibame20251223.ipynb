````python
import urllib.request as req
import pandas as pd

url = "https://github.com/Elwing-Chou/tibame20251203/raw/refs/heads/main/highrail.csv"
req.urlretrieve(url, "highrail.csv")
df = pd.read_csv("highrail.csv", encoding="utf-8", index_col=0)
df
````
