````python
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# print(type(content))
content = json.loads(content)
# print(type(content))
danmus = content["data"]["danmu"]

f = open("danmu.csv", "w", encoding="utf-8")
f.write("userid,text\n")
for d in danmus:
    d_uid = d["userid"]
    d_text = d["text"]
    # print(d_uid, d_text)
    pat = '{},"{}"\n'
    f.write(pat.format(d_uid, d_text))
f.close()
````

````python
import urllib.request as request
import urllib.parse as parse
import json
import pandas as pd
import datetime as dt
today = dt.date.today()
today_s = today.strftime("%Y/%m/%d")

stops = [
    "NanGang",
    "TaiPei",
    "BanQiao",
    "TaoYuan",
    "XinZhu",
    "MiaoLi",
    "TaiZhong",
    "ZhangHua",
    "YunLin",
    "JiaYi",
    "TaiNan",
    "ZuoYing",
]

hint = ""
for i in range(len(stops)):
    hint = hint + "[{}] {}\n".format(i, stops[i])
print("Ë´ãÈÅ∏Âá∫Áôº:")
start_stop_idx = int(input(hint))
start_stop = stops[start_stop_idx]
print("Ë´ãÈÅ∏Âà∞ÈÅî:")
end_stop_idx = int(input(hint))
end_stop = stops[end_stop_idx]

url = 'https://www.thsrc.com.tw/TimeTable/Search'
# Data to be sent (dictionary, list of tuples, etc.)
data = {
    "SearchType":"S",
    "Lang":"TW",
    "StartStation":start_stop,
    "EndStation":end_stop,
    "OutWardSearchDate":today_s,
    "OutWardSearchTime":"14:30",
    "ReturnSearchDate":today_s,
    "ReturnSearchTime":"14:30",
    "DiscountType":"",
}

# Encode the data into the standard format and then to bytes
encoded_data = parse.urlencode(data).encode('utf-8')
resp = request.urlopen(url, encoded_data)
content = resp.read()
content = json.loads(content)
departure = content["data"]["DepartureTable"]
departure_train = departure["TrainItem"]
df = pd.DataFrame(departure_train)
df
````
